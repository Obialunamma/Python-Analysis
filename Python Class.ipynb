```python
# Get the weight from the user as a float
weight = float(input("Weight: "))

# Get the unit type from the user (either 'K' for kilograms or 'L' for pounds)
unit = input("(K)g or (L)bs: ").strip().upper()

# Convert and print the weight based on the unit
if unit == "K":
    converted = weight / 0.453592
    print("Weight in Lbs: " + str(converted))
elif unit == "L":
    converted = weight * 0.453592
    print("Weight in Kgs: " + str(converted))
else:
    print("Invalid unit. Please enter 'K' for kilograms or 'L' for pounds.")
```

##### strftime() function
The datetime object has a method for formatting date objects into readable strings.
The method is called strftime(), and takes one parameter, format, to specify the format of the returned string:

**Directive**	       **Description**	                      **Example**
%a	                Weekday,short version	                     Wed	
%A	                Weekday,full version	                     Wednesday	
%w	                Weekday as a number 0-6,0 is Sunday	         3	
%d	                Day of month 01-31	                         31	
%b	                Month name, short version                    Dec	
%B	                Month name, full version	                 December	
%m	                Month as a number 01-12	                     12	
%y	               Year, short version, without century	         18	
%Y	               Year, full version	                         2018

```python
import datetime
x = datetime.datetime(2018, 6, 1)
print(x.strftime("%B"))
```

 ### Python Data Structures
 **Lists**
A list is an ordered collection of items. Lists are mutable, meaning you can change their contents.

```python
fruits.append('pawpaw')
print(fruits)
```

```python
# Example of a list
fruits = ['apple', 'banana', 'cherry']
print(fruits)

# Adding an item to a list
fruits.append('date')
print(fruits)

# Accessing elements by index
print(fruits[1]) 
```

**Dictionaries**
A dictionary stores data in key-value pairs. It's useful for looking up values by their keys.


```python
# Example of a dictionary
student = {'name': 'John', 'age': 22, 'major': 'Computer Science'}
print(student)


```

**Sets**
A set is an unordered collection of unique items.

```python
# Example of a set
numbers = {1, 2, 3, 4, 4, 5}
print(numbers)  # Duplicate '4' is removed

# Adding an element to a set
numbers.add(6)
print(numbers)
```

**Tuples**
A tuple is an ordered, immutable collection of items.

```python
# Example of a tuple
coordinates = (10.0, 20.0)
print(coordinates)

# Accessing elements by index
print(coordinates[0]) 
```

```python
import pandas as pd
import calendar

# Assuming 'walmart_sales' DataFrame is already available and has a 'Date' column of datetime type

# Extract month number from the 'Date' column and add it to the DataFrame
walmart_sales['Month'] = walmart_sales['Date'].dt.month

# Group by 'Month' and aggregate 'Weekly_Sales'
monthly_sales = walmart_sales.groupby(['Month'])['Weekly_Sales'].sum().reset_index()

# Rename the columns for clarity
monthly_sales.rename(columns={'Weekly_Sales': 'TotalSales'}, inplace=True)

# Map month numbers to month names
monthly_sales['Month'] = monthly_sales['Month'].apply(lambda x: calendar.month_name[x])

# Display the aggregated sales data with month names
print(monthly_sales.head())
```
